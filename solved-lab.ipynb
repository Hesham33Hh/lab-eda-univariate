### Categorías de productos

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import skew, kurtosis

# Cargar el archivo CSV
df = pd.read_csv("/Users/heshamelsharkawy/Documents/coding/Ironhack/week_11/day_2/amz_uk_price_prediction_dataset.csv")

# Ver las primeras filas
df.head()
```

```python
# Tabla de frecuencias
category_counts = df['product_category'].value_counts()
print("Top 5 categorías más cotizadas:\n", category_counts.head(5))
```

```python
print(df.columns)
```

```python
# Tabla de frecuencias
category_counts = df['category'].value_counts()
print("Top 5 categorías más cotizadas:\n", category_counts.head(5))
```

```python
top5_categories = category_counts.head(5)

import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x=top5_categories.values, y=top5_categories.index)
plt.title('Top 5 categorías más cotizadas en Amazon UK')
plt.xlabel('Frecuencia')
plt.ylabel('Categoría')
plt.show()
```

```python
top5_categories.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Proporción de las 5 principales categorías')
plt.ylabel('')
plt.show()
```

### Análisis de precios (price)

```python
print("Media:", df['price'].mean())
print("Mediana:", df['price'].median())
print("Moda:", df['price'].mode()[0])
```

```python
print("Varianza:", df['price'].var())
print("Desviación estándar:", df['price'].std())
print("Rango:", df['price'].max() - df['price'].min())
print("IQR:", df['price'].quantile(0.75) - df['price'].quantile(0.25))
```

```python
sns.boxplot(x=df['price'])
plt.title('Boxplot de precios')
plt.xlabel('Precio')
plt.show()
```

### Análisis de calificaciones (stars)

```python
print("Media:", df['stars'].mean())
print("Mediana:", df['stars'].median())
print("Moda:", df['stars'].mode()[0])
```

```python
print("Varianza:", df['stars'].var())
print("Desviación estándar:", df['stars'].std())
print("IQR:", df['stars'].quantile(0.75) - df['stars'].quantile(0.25))
```

```python
print("Asimetría (skew):", skew(df['stars'].dropna()))
print("Curtosis:", kurtosis(df['stars'].dropna()))
```

```python
sns.histplot(df['stars'], bins=10, kde=True)
plt.title('Distribución de calificaciones de productos')
plt.xlabel('Calificación (stars)')
plt.ylabel('Frecuencia')
plt.show()
```
